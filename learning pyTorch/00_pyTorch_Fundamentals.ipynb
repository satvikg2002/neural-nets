## Reshaping, stacking, squeezing and unsqueezing
torch.reshape(input, shape)	-  Reshapes input to shape (if compatible), can also use torch.Tensor.reshape().   
Tensor.view(shape) -  Returns a view of the original tensor in a different shape but shares the same data as the original tensor.  
torch.stack(tensors, dim=0)	-   Concatenates a sequence of tensors along a new dimension (dim), all tensors must be same size.  
torch.squeeze(input) - Squeezes input to remove all the dimensions with value 1.  
torch.unsqueeze(input, dim) -	  Returns input with a dimension value of 1 added at dim.   
torch.permute(input, dims) -  Returns a view of the original input with its dimensions permuted (rearranged) to dims.  

```python
import torch
x = torch.arange(1., 10.)
x, x.shape
```

```python
x_reshaped = x.reshape(9,1)
x_reshaped, x_reshaped.shape
```

```python
z = x.view(1,9)
z,z.shape
```

```python
z[:, 0] = 5
z,x
```

```python
x_stacked = torch.stack([x,x,x,x], dim=0)
x_stacked
```

```python
x_reshaped, x_reshaped.shape
```

```python
x_squeezed = x_reshaped.squeeze()
x_squeezed, x_squeezed.shape
```

```python
x_unsqueezed = x_squeezed.unsqueeze(dim=0)
x_unsqueezed, x_unsqueezed.shape
```

```python
x_col = torch.rand(size=(244, 244, 3))
x_col.shape
```

```python
x_perm = x_col.permute(2,0,1)
x_perm.shape
```

```python
x_col[0,0,0] = 100
x_perm[0,0,0]
```

## Reproducibility
NNs
start with random numbers -> tensor operations -> try to make better (again and again and again)  



```python
tensor_A = torch.rand(3,4)
tensor_B = torch.rand(3,4)

print("Tensor A :\n", tensor_A, "\n\nTensor B :\n", tensor_B, "\n\n")
tensor_A == tensor_B
```

```python
import random

RANDOM_SEED=42
torch.manual_seed(seed=RANDOM_SEED)
tensor_C = torch.rand(3,4)

# Have to reset seed everytime new rand() is called for reproducibility
torch.random.manual_seed(seed=RANDOM_SEED)
tensor_D = torch.rand(3,4)

print("Tensor C :\n", tensor_C, "\n\nTensor D :\n", tensor_D, "\n\n")
tensor_C == tensor_D
```
